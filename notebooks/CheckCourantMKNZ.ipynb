```python
import numpy as np
import numpy.ma as ma
import netCDF4 as nc
%matplotlib inline 
import matplotlib.pyplot as plt
from salishsea_tools import diagnosis_tools
import glob

src = '/ocean/imachuca/Canyons/results_mackenzie/idealized/Resolution_237_177_80/Run01/'
mesh_mask = nc.Dataset(src + 'mesh_mask.nc')
test_u = glob.glob(src + 'GYRE_1h*U*.nc')
test_v = glob.glob(src + 'GYRE_1h*V*.nc')
test_w = glob.glob(src + 'GYRE_1h*W*.nc')
```

```python
cu = diagnosis_tools.pcourantu(test_u, mesh_mask)
print("cu")
cv = diagnosis_tools.pcourantv(test_v, mesh_mask)
print("cv")
cw = diagnosis_tools.pcourantw(test_w, mesh_mask)
print("cw")
```

```python
plt.figure(figsize=(16,6))
dt=300
plt.subplot(1,3,1); plt.pcolormesh(cu*dt); plt.colorbar()
plt.subplot(1,3,2); plt.pcolormesh(cv*dt); plt.colorbar()
plt.subplot(1,3,3); plt.pcolormesh(cw*dt); plt.colorbar()
```
