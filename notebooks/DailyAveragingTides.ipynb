```python
import numpy as np
from IPython import embed
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
w=2*np.pi/44712   # M2 Tidal frequency
N=40 # number of days

plt.figure(figsize=(8,4))
plt.subplot(2,1,1)
ta=[0]*N
td=[0]*N
for i in range(N):
    tt = np.linspace(  i*86400, (i+1)*86400, 500)
    tide = np.sin(w*tt)
    plt.plot(tt/86400,tide,'r')
    td[i] = i + 0.5   # 
    ta[i] = np.mean(tide)
    
plt.subplot(2,1,2)
plt.plot(td,ta,'k')
plt.tight_layout(pad=0.5)

```
